```python
#plot the number of bookings per month
df['date_time'] = pd.to_datetime(df['date_time'])
df['month'] = df['date_time'].dt.month
df['year'] = df['date_time'].dt.year
df.groupby('month')['booking_bool'].sum().plot(kind='bar', figsize=(10,5))
```

```python
#print for each srch column the top 5 value counts 
for col in df.columns:
    if 'srch' in col:
        print(df[col].value_counts().head())
```

So, most searched length of stay is 1 night. Most people searched for a hotel the day before the start of their stay. Most searched for number of adults is 2, while most searched for number of children is 0. Most searced for number of rooms is 1. Whether the search included a Saturday night is approximately 50%. 

```python
#check whether booking_bool is 1 when random_bool is 0
df[df['random_bool'] == 0]['booking_bool'].value_counts()
```

So when the normal sort order was displayed, 12948 people did a booking. 


```python
df[df['random_bool'] == 1]['booking_bool'].value_counts()
```

While if the sort order is random, only 807 people did a booking.

```python
#at what time are most bookings made
df['hour'] = df['date_time'].dt.hour
df.groupby('hour')['booking_bool'].sum().plot(kind='bar', figsize=(10,5));
```

So most bookings between 11 and 18.

```python
#at what day of the week are most bookings made
df['day'] = df['date_time'].dt.dayofweek
df.groupby('day')['booking_bool'].sum().plot(kind='bar', figsize=(10,5))
plt.xticks(np.arange(7), ('Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'));
```

So most bookings at the beginning of the week, then it decreases.

```python
df['date_time']  = df['date_time'].apply(lambda x: (str(x)[:7]) if x == x else np.nan)

date_bookings  = df.groupby('date_time')["booking_bool"].sum()
ax1 = date_bookings.plot(legend=True,marker='o',title="Total Bookings per month", figsize=(15,5)) 
ax1.set_xticks(range(len(date_bookings)))
xlabels = ax1.set_xticklabels(date_bookings.index.tolist(), rotation=90)
```

```python
date_clicks  = df.groupby('date_time')["click_bool"].sum()
ax2 = date_clicks.plot(legend=True,marker='o',title="Total Clicks per month", figsize=(15,5))
ax2.set_xticks(range(len(date_clicks)))
xlabels = ax2.set_xticklabels(date_clicks.index.tolist(), rotation=90)
```
