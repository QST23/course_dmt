## Basic

```python
#delete column if type is string
def delete_string_columns(df):
    for col in df.columns:
        if df[col].dtype==pl.Utf8:
            df = df.drop(col)

    return df
df.head()
```

```python
#delete all columns which name ends with _id

def delete_id_columns(df):
    for col in df.columns:
        if col.endswith('_id'):
            df = df.drop(col)
    return df
df.head()
```

```python
#rescale all columns to [0,1]
def rescaler(df):
    for col in df.columns:
        # df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
        df = df.with_columns(
            (pl.col(col) - pl.col(col).min()) / (pl.col(col).max() - pl.col(col).min())
        )
    return df
df.head()
```

## Imputing

### Basic (mean everything)

```python
#find null values
print(sum(df.null_count()))

#find nan values
df.fill_null(strategy="mean")
```

```python
# fill all missing values with their mean
def impute_mean(df:pl.DataFrame)->pl.DataFrame:
    # Calculate the mean of each column
    means = df.mean()

    #loop over the columns of mean
    for col in means:
        # Fill in missing values with their respective column means
        df = df.with_columns(
            pl.when(pl.is_nan(col)).then(means[col]).otherwise(col).alias(col)
        )
    
    # # Fill in missing values with their respective column means
    # imputed_data = df.fill_nan(means)
    
    return df

impute_mean(df)
```

```python
def impute_mean_values(df:pl.DataFrame, identifier, feature_name):
    # Group by user ID and feature, and calculate the mean of the value column
    grouped_data = df.groupby([identifier, feature_name]).select([identifier, feature_name,]).mean()
    # grouped_data = grouped_data. ("value", "imputed_value")])
    
    # Join the imputed values back into the original dataframe
    imputed_data = df.join(grouped_data, on=[identifier, feature_name], how="left")
    
    # Fill in missing values with the imputed values, and drop the imputed_value column
    imputed_data["value"] = imputed_data["value"].fillna(imputed_data["imputed_value"])
    imputed_data = imputed_data.drop("imputed_value")
    
    return imputed_data

impute_mean_values(df, "visitor_hist_adr_usd")
```
