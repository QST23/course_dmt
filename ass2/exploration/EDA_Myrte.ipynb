```python
#for statistics, skip all id columns and date_time
useful_columns = df.drop(['srch_id', 'site_id', 'visitor_location_country_id','date_time', 'prop_id', 'prop_country_id', 'prop_brand_bool', 'promotion_flag', 'srch_destination_id', 'srch_saturday_night_bool', 'random_bool', 'click_bool', 'booking_bool'], axis=1)
#also skip all columns with competition_rate
useful_columns = useful_columns.drop(['comp1_rate', 'comp1_inv', 'comp1_rate_percent_diff', 'comp2_rate', 'comp2_inv', 'comp2_rate_percent_diff', 'comp3_rate', 'comp3_inv', 'comp3_rate_percent_diff', 'comp4_rate', 'comp4_inv', 'comp4_rate_percent_diff', 'comp5_rate', 'comp5_inv', 'comp5_rate_percent_diff', 'comp6_rate', 'comp6_inv', 'comp6_rate_percent_diff', 'comp7_rate', 'comp7_inv', 'comp7_rate_percent_diff', 'comp8_rate', 'comp8_inv', 'comp8_rate_percent_diff'], axis=1)

#make table with missing values, mean, std, min, 25%, 50%, 75%, max per column of useful_columns, not using describe because it skips missing values
#initialize dataframe
properties = pd.DataFrame()
#add missing values per attribute
properties['missing'] = useful_columns.isnull().sum()
properties['mean'] = useful_columns.mean()
properties['std'] = useful_columns.std()
properties['min'] = useful_columns.min()
properties['25%'] = useful_columns.quantile(0.25)
properties['50%'] = useful_columns.quantile(0.5)
properties['75%'] = useful_columns.quantile(0.75)
properties['max'] = useful_columns.max()
#add number of records per attribute

#round all values to 2 decimals
properties = properties.round(2)
properties
#convert dataframe to table
properties.to_csv('properties.csv')
```

So most bookings at the beginning of the week, then it decreases.
