```python
#which variables do have missing values and how many
df.loc[df['value'].isnull()]['variable'].value_counts()

#which ids do have missing values
df.loc[df['value'].isnull()]['id'].unique()

#inspect missing values for id 01
df.loc[df['id'] == 'AS14.01'].loc[df['value'].isnull()]
```

```python
#find outliers
def find_outliers(series):
    outliers = []
    std = np.std(series)
    threshold = 3 * std
    mean = np.mean(series)
    
    for y in series:
        value = (y - mean)
        if np.abs(value) > threshold:
            outliers.append(y)

    return outliers

for i in df['variable'].unique():
    outliers = find_outliers(df.loc[df['variable'] == i]['value'])
    print('Outliers for {}: {}'.format(i, len(outliers)))
```
